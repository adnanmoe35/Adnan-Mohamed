```python
# I will create a DataFrame with the following columns: Name, Age, City, Income Include at least two
# missing values (e.g., None or NaN). 2. Check which values are missing. 3. Fill missing ages with
# the average age. 4. Fill missing income with 0. 5. Create a new column: Income_Level = "High"
# if Income > 60000 else "Low" 6. Display only the rows where Income_Level is "High".

import pandas as pd
import numpy as np

# 1. Create DataFrame with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, np.nan, 30, 35, np.nan],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney'],
    'Income': [70000, 50000, np.nan, 80000, np.nan]
}
df = pd.DataFrame(data)

# 2. Check which values are missing
print("Missing values per column:\n", df.isnull().sum())

# 3. Fill missing ages with the average age
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 4. Fill missing income with 0
df['Income'] = df['Income'].fillna(0)

# 5. Create Income_Level column
df['Income_Level'] = df['Income'].apply(lambda x: 'High' if x > 60000 else 'Low')

# 6. Display only "High" income rows
high_income_df = df[df['Income_Level'] == 'High']
print("\nHigh Income Rows:\n", high_income_df)
```

```python
# Create a DataFrame with: Product, Category, Price, Quantity 2. Compute a new column:
# Total_Value =Price * Quantity 3. Group by Category and compute: o Total quantity sold o Average
# price o Total value 4. Sort categories by Total_Value in descending order. 5. Display the top-
# earning category.

# 1. Create DataFrame
sales_data = {
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Laptop', 'Monitor'],
    'Category': ['Electronics', 'Peripherals', 'Electronics', 'Peripherals', 'Electronics', 'Electronics'],
    'Price': [1200, 25, 300, 75, 1200, 300],
    'Quantity': [5, 20, 10, 15, 3, 7]
}
df_sales = pd.DataFrame(sales_data)

# 2. Compute Total_Value column
df_sales['Total_Value'] = df_sales['Price'] * df_sales['Quantity']

# 3. Group by Category and compute metrics
category_summary = df_sales.groupby('Category').agg(
    Total_Quantity=('Quantity', 'sum'),
    Average_Price=('Price', 'mean'),
    Total_Value=('Total_Value', 'sum')
)

# 4. Sort categories by Total_Value in descending order
category_summary = category_summary.sort_values(by='Total_Value', ascending=False)

# 5. Display the top-earning category
print("Category Summary:\n", category_summary)
print("\nTop Earning Category:", category_summary.index[0])
```

```python
# Category Level Sales Analysis (Working with pandas data structure)


# 1) Group by Product and calculate total units sold and revenue
product_summary = df_sales.groupby("Product").agg(
    Total_Units=("Quantity", "sum"),
    Total_Revenue=("Total_Value", "sum")
).reset_index()

print("Product Summary:")
display(product_summary)

# 2) Sort products by revenue (highest first)
product_sorted = product_summary.sort_values("Total_Revenue", ascending=False)

print("Products sorted by revenue:")
display(product_sorted)

# 3) Identify the top selling product
top_product = product_sorted.iloc[0]

print(f"Top selling product: {top_product['Product']}")
print(f"Total revenue generated: ${top_product['Total_Revenue']}")
```

- What we have here is a deeper analysis to the previous code completed above using GroupBy

- The product generating the highest revenue is Laptop ($9600).

- Sorting the products helps identify the best performing items.

- Even though Mouse sold the most units, it generated the lowest revenue, showing that price strongly affects revenue.

```python
# Working with Messy Text files in Pandas (Student Exercise).

# Working with messy data

# List the messy data
messy_text = """
community | participants | avg_age | device_access | notes
Kijiji | 23 | 31 | mobile phones | ; tablets | inconsistant attendance 
Makondo|17| | mobile phones| missing age data
Luwero | 45|29 mobile phones ; laptops | 
Gulu| | 34 | tablets | device count unclear 
Mbale | 38 | 28| mobile phones ; laptops; tablets | good engagement
"""
print("Messy Text Data:\n", messy_text)
```

```python
# Load messy data text file
import pandas as pd
from io import StringIO
df_raw = pd.read_csv(
    StringIO(messy_text),
    sep="\\|",
    engine="python",
    skip_blank_lines=True
)
df_raw
```

- After loading the messy text data into a DataFrame, we can see that the structure is not perfectly clean.

- Some rows contain missing values, inconsistent formatting, or misplaced data.  

- These types of issues are common in real-world datasets and often require additional data cleaning steps before analysis.

- Despite the messy formatting, Pandas allows us to quickly convert the text data into a structured table that can be explored and cleaned further.

```python
# Section 1 (ICT4D- Text Data, from the same topic above). 

df = df_raw.copy()

# 1) Clean column names and strip spaces
df.columns = df.columns.str.strip()
for c in df.columns:
    df[c] = df[c].astype(str).str.strip()

# 2) Fix participants and avg age to numeric, coercing errors to NaN
df["participants"] = pd.to_numeric(df["participants"], errors="coerce")
df["avg_age"] = pd.to_numeric(df["avg_age"], errors="coerce")

# Fill missing avg_age with overall mean age (ignoring NaN)
df["avg_age"] = df["avg_age"].fillna(df["avg_age"].mean())

# 3) Basic ICT4D features from device_access text
df["device_access"] = df["device_access"].replace("None", "")
df["has_mobile"] = df["device_access"].str.lower().str.contains("mobile|phone", na=False)
df["has_tablet"] = df["device_access"].str.lower().str.contains("tablet", na=False)
df["has_laptop"] = df["device_access"].str.lower().str.contains("laptop", na=False)

print("Device access counts (how many communities mention each):")
print("Mobile:", df["has_mobile"].sum())
print("Tablet:", df["has_tablet"].sum())
print("Laptop:", df["has_laptop"].sum())
```

```python
# Now we will view some quick insights based on the cleaned data
print("Top communities by participants:")
display(df.sort_values("participants", ascending=False)[["community", "participants"]])

# Finally, we can also check for any notes that indicate data issues

print("Rows with data issues (based on notes):")
df["notes"] = df["notes"].replace("None", "")
issues = df[df["notes"].str.lower().str.contains("missing|unclear|inconsistent", na=False)]
display(issues[["community", "notes"]])
```
